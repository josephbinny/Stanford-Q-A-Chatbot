```python
def extract_df(data_dict):
    df = pd.DataFrame()

    for title_paragraph in data_dict["data"]:
        title = title_paragraph["title"]
        for paragraph in title_paragraph['paragraphs']:
            for QNA in paragraph['qas']:
                answers = []
                answers_start = []
                for answer in QNA["answers"]:
                    answers.append(answer["text"])
                    answers_start.append(answer["answer_start"])
                question = [QNA["question"]] * len(answers)
                context = [paragraph["context"]] * len(answers)
                titles = [title] * len(answers)
                df_curr = pd.DataFrame({"title": titles,
                                        "context": context,
                                        "question": question,
                                        "answer": answers,
                                        "answer_start": answers_start})
                df = pd.concat([df, df_curr])
    return df
```
